# 01 EDA and Preprocessing\n
Loads DS1 from UCI, DS2 and DS3 from local CSV paths, then builds merged engineered dataset.

```python
import json\n
from pathlib import Path\n
import sys\n
\n
sys.path.append(str(Path('..') / 'src'))\n
\n
from maternaguard_ml.pipeline import build_merged_dataset, save_inspection
```

```python
DS2_PATH = '../data/raw/ds2.csv'\n
DS3_PATH = '../data/raw/ds3.csv'\n
\n
merged, inspection = build_merged_dataset(DS2_PATH, DS3_PATH)\n
Path('../data/processed').mkdir(parents=True, exist_ok=True)\n
Path('../docs').mkdir(parents=True, exist_ok=True)\n
merged.to_csv('../data/processed/merged_dataset.csv', index=False)\n
save_inspection(inspection, '../docs/dataset_inspection.json')\n
print('Merged shape:', merged.shape)\n
print('Class distribution:', merged['risk_level'].value_counts().sort_index().to_dict())\n
print('Inspection saved to ../docs/dataset_inspection.json')
```
