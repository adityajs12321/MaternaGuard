# 03 Hybrid Ensemble\n
Trains ANN + RF and evaluates RF, ANN, and soft-voting ensemble.

```python
from pathlib import Path\n
import pandas as pd\n
import sys\n
\n
sys.path.append(str(Path('..') / 'src'))\n
\n
from maternaguard_ml.pipeline import split_scale_and_balance\n
from maternaguard_ml.training import train_random_forest, train_ann, evaluate_all, save_artifacts, print_metrics
```

```python
merged = pd.read_csv('../data/processed/merged_dataset.csv')\n
prep = split_scale_and_balance(merged, models_dir='../models')\n
rf = train_random_forest(prep['X_train_scaled'], prep['y_train_sm'])\n
ann = train_ann(prep['X_train_scaled'], prep['y_train_sm'])\n
metrics = evaluate_all(rf, ann, prep['X_test_scaled'], prep['y_test'])\n
print_metrics(metrics)\n
save_artifacts(rf, ann, metrics, models_dir='../models', docs_dir='../docs')
```
